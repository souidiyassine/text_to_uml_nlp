```python
with open('dataset.txt') as f:
    lines = f.readlines()
```

```python
lines
```

```python
reqs = []
for line in lines:
    if line.startswith("F:"):
        reqs.append(line.replace("F:", "").strip())
```

```python
reqs[0]
```

```python
len(reqs)
```

```python
import nltk
```

```python
from functions import *
```

```python
labels = []
for req in reqs:
    if not req.endswith("."):
        req += "."
    text_token = nltk.sent_tokenize(str(req))
    for token in text_token:
        wordsList = nltk.word_tokenize(token)
    wordsList
    l = []
    uml, inheritance, relationship, object, object_inh = text_to_uml(str(req))
    for w in wordsList:
        e, a, r = 0, 0, 0
        for entities, attr in uml.items():
            if w == entities:
                e = 1
                break
            if len(attr) > 0:
                if w == attr[0][0]:
                    a = 1
                    break
        if w in relationship:
            r = 1
        if e == 0 and a == 0 and r == 0:
            l.append("O")
        elif e == 1:
            l.append("entity")
        elif a == 1:
            l.append("attr")
        elif r == 1:
            l.append("relationship")
    l = " ".join(l)
    labels.append(l)
```

```python
len(labels)
```

```python
import pandas as pd
df = pd.DataFrame({'req': reqs, 'labels': labels})
df
```

```python
df.to_csv("ner_dataset.csv", index=False)
```
